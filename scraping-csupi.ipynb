#### Scraping Instagram

```python
#import library
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from selenium.webdriver import Chrome
from instascrape import Profile, scrape_posts
```

```python
# membuat jalur untuk mengakses webdriver
PATH = r"C:\Windows\chromedriver.exe"
webdriver = Chrome(PATH)

# Scraping computer sceince upi profile
SESSIONID = 'ENTER_YOUR_SESSION_ID_HERE'
headers = {"user-agent": "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Mobile Safari/537.36 Edg/87.0.664.57",
           "cookie": f"sessionid={SESSIONID};"}
cs= Profile("computerscienceupi")
cs.scrape(headers=headers)

# Scraping computer science upi posts
posts = cs.get_posts(webdriver=webdriver, login_first=True)
scraped, unscraped = scrape_posts(posts, silent=False, headers=headers, pause=10)
```

```python
#memasukan hasil scrapinng ke data frame menggunakan pandas
posts_data = [post.to_dict() for post in scraped]
posts_df = pd.DataFrame(posts_data)
print(posts_df[['id', 'upload_date', 'comments', 'likes', 'shortcode']])
```

##### Processing Data

```python
#melihat data kosong
posts_df.isnull().sum()
```

```python
#melihat persentase atribut kosong
df_na = [a for a in posts_df.columns if posts_df[a].isna().sum()>0]
for a in df_na:
    print(a, np.round(posts_df[a].isna().mean()*100,4), '%')
```

```python
#visualisasi data kosong
import matplotlib.pylab
matplotlib.pylab.rcParams['figure.figsize'] = (8, 6)
sns.heatmap(posts_df.isnull(), yticklabels=False, cbar=False, cmap='inferno')
```

```python
df1 = posts_df.copy()
```
